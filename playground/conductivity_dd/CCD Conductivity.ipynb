```python
%matplotlib inline
import numpy as np
import matplotlib as mpl
import pylab as plt
```

```python
def ccd_complex(f, pars):
    sigma_infty = pars[0]
    m = pars[1]
    tau = pars[2]
    c = pars[3]
    
    omega = 2 * np.pi * f
    
    sigma_complex = sigma_infty * (1 - m / (1 + (1j * omega * tau) ** c))
    return sigma_complex
```

```python
f = np.logspace(-3, 3, 30)
pars = [1, 0.1, 0.4, 0.9]
erg = ccd_complex(f, pars)

fig, axes = plt.subplots(2, 1, figsize=(7, 4))
ax = axes[0]
ax.semilogx(f, np.real(erg), '.-')

ax = axes[1]
ax.semilogx(f, np.imag(erg), '.-')
```

```python
def sig_imag(f, pars):
    sigma_infty = pars[0]
    m = pars[1]
    tau = pars[2]
    c = pars[3]
    
    omega = 2 * np.pi * f
    owt = (omega * tau) ** (2 * c)
    angcos = np.cos(c * np.pi / 2.0)
    denom = 1.0 + 2 * (omega * tau) ** c * angcos + owt

    sig_imag = sigma_infty * (m * (omega * tau) ** c * np.sin(c * np.pi / 2.0)) / denom
    return sig_imag
```

```python
f = np.logspace(-3, 3, 30)
pars = [1, 0.1, 0.4, 0.9]
erg = ccd_complex(f, pars)
erg_imag = sig_imag(f, pars)

fig, ax = plt.subplots(1, 1, figsize=(7, 4))
ax.semilogx(f, np.imag(erg), '.-')
ax.semilogx(f, erg_imag, '-', color='red')
```
